```python
import statsmodels.api as sm #linReg using statsmodel
```

```python
from sklearn import datasets,linear_model
data = datasets.load_boston() #importing the dataset
```

```python
import pandas as pd #importing pandas
```

```python
df = pd.DataFrame(data.data, columns=data.feature_names)
out = pd.DataFrame(data.target, columns=["MEDV"]) #splitting the feature and taget data
```

```python
x = df
y = out["MEDV"]
x = sm.add_constant(x) #adding an intercept constant
```

```python
model = sm.OLS(y,x).fit() #fitting the model
```

```python
pred = model.predict(x) #prediction
```

```python
model.summary() #describes the model
```
